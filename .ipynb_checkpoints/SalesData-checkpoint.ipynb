## Merge the 24 months of sales data into a one CSV file

```python
# create a list for all the data in the folder
files = [file for file in os.listdir(r'data')] 
#  create a pandas dataFrame
all_months_data = pd.DataFrame()
# create a loop for marge the data
for file in files:
    data = pd.read_csv("data/" + file)
    all_months_data = pd.concat([all_months_data, data])
# convert all data to csv    
all_months_data.to_csv("sales_data.csv", index=False)
```

## Preprocessing Data


```python
# drop the rows of Null data
sales_data = sales_data.dropna(how='all')
sales_data[sales_data.isna().any(axis=1)]
sales_data = sales_data[sales_data['Order Date'].str[0:2] != 'Or']
sales_data.head()
```

```python
sales_data.describe()
```

```python
#convert Order Date to datetime using to_datetime()
sales_data['Order Date'] = pd.to_datetime(sales_data['Order Date'])
sales_data
```

```python
# convert Quantity Ordered to int using astype
sales_data['Quantity Ordered'] = sales_data['Quantity Ordered'].astype('int64')
```

```python
#convert Price Each to float using astype
sales_data['Price Each'] =sales_data['Price Each'].astype('float')
```

```python
# check data info to know the type of Column
sales_data.info()
```

## Exploratory data analysis

```python
#make new columns contain year, month, day and hour from Order Date column
sales_data['Year'] =sales_data['Order Date'].dt.year
sales_data['Month'] = sales_data['Order Date'].dt.month
sales_data['day']  = sales_data['Order Date'].dt.day
sales_data['Hour'] = sales_data['Order Date'].dt.hour
```

```python
#make new column contain only address from Purchase Address
sales_data['Addr'] = sales_data['Purchase Address'].str.split(' ').apply(lambda x:x[-2])
```

```python
#decribe Revenue for each row
sales_data['Sales'] = sales_data['Quantity Ordered'] * sales_data['Price Each'] 
```

# Data Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
df = pd.DataFrame(sales_data)  
df=df.drop(labels=['Order ID',"Order Date",'Purchase Address'],axis=1)#drop unnecessary columns
df
```

```python
# encoder object and int datatype
encoder = LabelEncoder() 

for i in df.columns: # Make a for loops
    if df[i].dtype == 'object': 
        encoder.fit_transform(list(df[i].values)) # Fit transform
        df[i] = encoder.transform(df[i].values) # Transform
         
        for j in df.columns: # Make a for loops again
            if df[j].dtype == 'int':
                df[j] = df[j].astype('float64') # Change the type
```

```python
#seplit data in data and lable 
X = df.drop(columns=[ 'Sales']) # Data X
y = df['Sales'] # Data y     
```

```python
#seplit data in train and validation
X_train, X_valid, y_train, y_valid = train_test_split(
    X, y,test_size=0.3, #select test size
    random_state=42)
```

```python
#using Random Forest to fit the model
RF = RandomForestRegressor()
RF.fit(X_train,y_train)#fit model
```

```python
# make a predict
pred = RF.predict(X_valid) 
```

```python
rf_r2 = RF.score(X_valid,y_valid)
print("R2 Score: ",RF.score(X_valid,y_valid))
print('MSE:{}'.format(mean_squared_error(y_valid,pred)))
```
